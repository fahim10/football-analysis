```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
clubs_data = pd.read_csv('/kaggle/input/player-scores/clubs.csv')

# Setting up the visualisation style
sns.set(style="whitegrid")

# Plotting the distribution of squad sizes
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
sns.histplot(clubs_data['squad_size'], bins=20, kde=True)
plt.title('Distribution of Squad Sizes')

# Plotting the distribution of average ages
plt.subplot(1, 3, 2)
sns.histplot(clubs_data['average_age'], bins=20, kde=True)
plt.title('Distribution of Average Ages')

# Plotting the distribution of number of foreigners
plt.subplot(1, 3, 3)
sns.histplot(clubs_data['foreigners_number'], bins=20, kde=True)
plt.title('Distribution of Number of Foreigners')

plt.tight_layout()
plt.show()
```
